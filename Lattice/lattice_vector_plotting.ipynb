```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from collections import defaultdict
from itertools import combinations, chain
import functools
import operator
import copy

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:70% !important; }</style>"))
```

```python
vector_arrays = defaultdict(list)
for i in [1,2,3]:
    with open('Site'+str(i)+'-KSRev-smallz.txt', 'r') as f:
        line = [[float(num) for num in fileline.split()] for fileline in f]
        vector_arrays[i] = line
```

Column 1 is $\frac{t}{U_0}$, while column 2 is $\frac{C}{U_0}$. New text files contain energies in the last column. Code will break since the zip merges all columns together. Need to remove the energy column to ensure correct coding behavior. 

```python
new_dict = defaultdict(list)
for i in [1, 2, 3]:
    for combo in combinations('xyz', 2):
        if i not in new_dict.keys():
            new_dict[i] = {functools.reduce(operator.add, combo) : vector_arrays[i]}
        else:
            new_dict[i].update({functools.reduce(operator.add, combo) : vector_arrays[i]})
```

```python
x, y, u1, v1, w1 = zip(*shortened_arrays[1]['xy'])
x, y, u2, v2, w2 = zip(*shortened_arrays[2]['xy'])
x, y, u3, v3, w3 = zip(*shortened_arrays[3]['xy'])
```

```python
u4 = np.add(np.add(u1, u2), u3) 
v4 = np.add(np.add(v1, v2), v3) 
w4 = np.add(np.add(w1, w2), w3) 

m4 = np.sqrt(np.power(u4, 2) + np.power(v4, 2) + np.power(w4, 2) )
```

```python
step_size = np.round(np.abs(y[0] - y[1]), 2)
```

```python
t_axis = np.arange(np.min(x), np.max(x) + step_size, step_size)
c_axis = np.arange(np.min(y) - step_size, np.max(y) + step_size, step_size)
```

```python
mag_array = np.reshape(m4, [len(t_axis), len(c_axis)])
u_array = np.reshape(u4, [len(t_axis), len(c_axis)])
v_array = np.reshape(v4, [len(t_axis), len(c_axis)])
w_array = np.reshape(w4, [len(t_axis), len(c_axis)])
```

```python
direction_dict = {}
array_dict = {}
for i, direction in enumerate(['x', 'y', 'z', '|m|']):
    direction_dict[i] = direction
    
for i, array in enumerate([u_array, v_array, w_array, mag_array]):
    array_dict[i] = array
```

```python
np.max(array_dict[2])
```

```python
# fig = plt.figure(figsize=[20,15])
# ax = fig.add_subplot(111)

for i in range(4):
    fig, axs = plt.subplots(1, figsize=[15, 10])
    vector_plot = axs.contourf(t_axis, c_axis, array_dict[i].T, levels=50)

    if i < 3:
        cbar = fig.colorbar(vector_plot, ax=axs).set_label('m_' + direction_dict[i])
    else:
        cbar = fig.colorbar(vector_plot, ax=axs).set_label(direction_dict[i])

    axs.set_ylabel('C/U_0')
    axs.set_xlabel('t/U_0')
    
    
    plt.savefig('Trimer_heatmap_' + direction_dict[i] + '.png')
    
    
plt.show()

```
